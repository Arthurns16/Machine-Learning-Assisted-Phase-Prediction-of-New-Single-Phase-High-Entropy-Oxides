```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import KFold,ShuffleSplit
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn import metrics
from sklearn.metrics import precision_score
import statistics
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Carrega os dados do arquivo Excel e filtra para deixar apenas as 243 features
dataset = pd.read_excel('Input_Total.xlsx')
dataset = dataset.loc[:,dataset.apply(pd.Series.nunique) != 1]
dataset = dataset.filter(regex=r'^(?!.*Keq)')
dataset = dataset.drop('Classe', axis=1)
dataset = dataset.drop('Átomos', axis=1)
dataset = dataset.drop('DOI', axis=1)
print(dataset.columns)


```

```python

X = dataset
dataset2 = pd.read_excel('Input_Total.xlsx')
y = dataset2['Classe']
print(y)
```

```python
print(X)
print(y)
```

```python
# Divide o conjunto de dados em X e y


scaler = StandardScaler()
scaler.fit(X)
X= scaler.transform(X)

#Divide os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)




```

```python
#X_train.reset_index(drop=True, inplace=True)

y_test = np.array(y_test)
y_train = np.array(y_train)

print(y_test)
print(y_train)
```

```python
import pickle
#Remember the time BP

if 1<2 :
  with open('X_trainBP.pickle', 'wb') as f:
    pickle.dump(X_train, f)

  with open('X_testBP.pickle', 'wb') as f:
    pickle.dump(X_test, f)

  with open('y_trainBP.pickle', 'wb') as f:
    pickle.dump(y_train, f)

  with open('y_testBP.pickle', 'wb') as f:
    pickle.dump(y_test, f)

  with open('X_BP.pickle', 'wb') as f:
    pickle.dump(X, f)

  with open('y_BP.pickle', 'wb') as f:
    pickle.dump(y, f)

  with open('scalerBP.pickle', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
################################################################################# Tuning hiperparameters

param_grid = {
      'hidden_layer_sizes':  [ [30], [36], [40], [47], [30,30], [36,36], [40,40], [47,47], [55,55] ],
      'activation' : ['identity', 'logistic', 'tanh', 'relu'],
       'solver' : ['lbfgs', 'sgd', 'adam'],
       'alpha' : [0.0001, 1e-16, 0.1],
       'max_iter' : [20000],
       'learning_rate_init' : [0.1, 0.001,0.000001],
       'learning_rate' : ['constant', 'invscaling', 'adaptive']

  }

classifier = MLPClassifier()
grid_search = GridSearchCV(estimator = classifier, param_grid = param_grid,
                            cv = 2, n_jobs = -1, verbose = 2)



grid_search.fit(X_train, y_train)
print(grid_search.best_params_)



#{'activation': 'relu', 'alpha': 0.0001, 'hidden_layer_sizes': [36, 36], 'learning_rate': 'constant', 'learning_rate_init': 0.1, 'max_iter': 20000, 'solver': 'lbfgs'}
#################################################################################
```

```python
import pickle

# Função para carregar um arquivo pickle
def load_pickle(file_name):
    with open(file_name, 'rb') as f:
        return pickle.load(f)

# Carregar os dados salvos
X_train = load_pickle('X_trainBP.pickle')
X_test = load_pickle('X_testBP.pickle')
y_train = load_pickle('y_trainBP.pickle')
y_test = load_pickle('y_testBP.pickle')
X = load_pickle('X_BP.pickle')
y = load_pickle('y_BP.pickle')
scaler = load_pickle('scalerBP.pickle')

# Exemplo de como verificar se os dados foram carregados corretamente
print("X_train:", X_train)
print("X_test:", X_test)
print("y_train:", y_train)
print("y_test:", y_test)
print("X:", X)
print("y:", y)
print("Scaler:", scaler)
```

```python
l=0
a=0
while l< 1000:
  model = MLPClassifier(learning_rate_init=0.001,activation='relu',
                        hidden_layer_sizes= [55, 55], alpha=  0.0001,
                        learning_rate= 'adaptive', solver= 'sgd',max_iter=200000)



  #{'activation': 'relu', 'alpha': 1e-16, 'hidden_layer_sizes': [36], 'learning_rate': 'invscaling',
  # 'learning_rate_init': 0.1, 'max_iter': 20000, 'solver': 'sgd'}

  model.fit(X_train,y_train)


  x = accuracy_score(y_test, model.predict(X_test), sample_weight=None)
  print(float(x))

  if x > a:
    a = x
    k = model


  l = l +1
```
